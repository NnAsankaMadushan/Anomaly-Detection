```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# # Assuming 'dataset' is your DataFrame
# # Initialize the StandardScaler
# standard_scaler = StandardScaler()

# # Fit and transform the data
# dataset = pd.DataFrame(standard_scaler.fit_transform(dataset), columns=dataset.columns)




from sklearn.preprocessing import MinMaxScaler
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
dataset = pd.DataFrame(min_max_scaler.fit_transform(dataset), columns=dataset.columns)
```

```python
x= dataset.drop('Label',axis=1)
y= dataset['Label']
```

```python
from sklearn.model_selection import train_test_split
from collections import Counter

# # First, split the dataset into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# # Then, split the training data further into training and validation sets
# # x_train, x_validation, y_train, y_validation = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# #Check the number of records
# print('The number of records in the training dataset is', x_train.shape[0])
# print('The number of records in the test dataset is', x_test.shape[0])

# # Analyze class distribution in the training set
# class_counts = Counter(y_train)
# majority_class, majority_count = max(class_counts.items(), key=lambda x: x[1])
# minority_class, minority_count = min(class_counts.items(), key=lambda x: x[-1])

# print(f"The training dataset has {majority_count} records for the majority class ({majority_class})")
# print(f"and {minority_count} records for the minority class ({minority_class})")

```

```python
import numpy as np
from sklearn.svm import OneClassSVM
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings("ignore")

# Define parameter grid
param_grid = {
    'kernel': ['rbf', 'sigmoid'],
    'gamma': ['scale'],
    'nu': [0.05]  # nu is like a regularization parameter for One-Class SVM
}

# Perform GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(OneClassSVM(), param_grid, cv=3, scoring='accuracy')
grid_search.fit(x_train)

print("Best parameters found:", grid_search.best_params_)

# Get the best One-Class SVM model
best_oc_svm = grid_search.best_estimator_

# Fit the model to training data
best_oc_svm.fit(x_train)

# Predict training and test data using One-Class SVM
y_train_pred = best_oc_svm.predict(x_train)  # 1 for normal, -1 for anomaly
y_test_pred = best_oc_svm.predict(x_test)

# Decision function to get anomaly scores (distance from the separating hyperplane)
train_scores = best_oc_svm.decision_function(x_train)
test_scores = best_oc_svm.decision_function(x_test)

# Set threshold based on training data scores (since x_train is only benign)
threshold = np.percentile(train_scores, 95)  # 95th percentile threshold

# Classify as anomalies or normal based on threshold
train_anomalies = train_scores < threshold  # True if score is less than threshold (anomaly)
test_anomalies = test_scores < threshold

# Count anomalies in training and test sets
print(f"Number of anomalies detected in training data: {np.sum(train_anomalies)}")
print(f"Number of anomalies detected in test data: {np.sum(test_anomalies)}")

# Optional: calculate accuracy, precision, or recall using actual labels if available
# y_train_true, y_test_true can be used if available
# accuracy_score(y_train_true, y_train_pred), etc.
```
