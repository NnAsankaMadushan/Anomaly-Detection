```python
import pandas as pd
import numpy as np
import ipaddress
import matplotlib.pyplot as plt

# Assuming the model is preloaded
# model = <Your trained model>

# Function to preprocess new data
def preprocess_new_data(csv_file):
    # Assuming similar preprocessing steps as your training data
    x_test = test_data.values
    timesteps = 1  # Ensure timesteps match the model's input shape
    x_test_reshaped = np.reshape(x_test, (x_test.shape[0], timesteps, x_test.shape[1]))
    
    return x_test_reshaped

# Function to calculate the dynamic threshold based on training/validation data
def calculate_threshold(reconstruction_errors):
    mean_re = np.mean(reconstruction_errors)
    std_re = np.std(reconstruction_errors)
    # threshold = mean_re + 3 * std_re
    threshold = np.percentile(mean_re, 95)
    return threshold

# Function to detect anomalies and visualize histogram
def detect_anomalies(csv_file):
    # Preprocess the new data
    x_test_reshaped = preprocess_new_data(csv_file)

    # Predict using the loaded model
    test_reconstructions = model.predict(x_test_reshaped)
    test_mse = np.mean(np.power(x_test_reshaped - test_reconstructions, 2), axis=(1, 2))

    # Calculate dynamic threshold based on reconstruction errors
    threshold = calculate_threshold(test_mse)

    # Predict anomaly labels based on the threshold
    y_pred = (test_mse > threshold).astype(int)  # 1 = Anomaly, 0 = Normal

    # Visualize the reconstruction error histogram
    visualize_reconstruction_error(test_mse, threshold)

    # Return the predicted anomalies
    return y_pred

# Function to visualize reconstruction error histogram
def visualize_reconstruction_error(test_mse, threshold):
    # Number of anomalies detected
    num_anomalies = np.sum(test_mse > threshold)

    plt.figure(figsize=(10, 6))
    
    # Create a histogram of reconstruction errors
    plt.hist(test_mse, bins=50, color='blue', edgecolor='black')
    
    # Add a vertical line to mark the threshold
    plt.axvline(x=threshold, color='r', linestyle='--', label=f"Threshold: {threshold:.2f}")
    
    # Add title and labels
    plt.title("Reconstruction Error Histogram")
    plt.xlabel("Reconstruction error")
    plt.ylabel("Number of samples")
    
    # Add text displaying the number of anomalies detected
    plt.text(threshold * 1.1, plt.ylim()[1] * 0.9, 
             f"Number of anomalies detected: {num_anomalies}",
             color='red', fontsize=12)
    
    # Show the plot with the legend
    plt.legend()
    plt.show()

# Example usage
csv_file = "/Users/M S I/Desktop/check/benign.csv"
anomalies = detect_anomalies(csv_file)
```
