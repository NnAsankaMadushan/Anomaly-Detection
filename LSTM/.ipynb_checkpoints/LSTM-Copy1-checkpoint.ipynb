```python
x= dataset.drop('Label',axis=1)
y= dataset['Label']
```

```python
from sklearn.model_selection import train_test_split

# First, split the dataset into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Then, split the training data further into training and validation sets
# x_train, x_validation, y_train, y_validation = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np

# Assuming x_train is a 2D array, reshape it to 3D for LSTM input
timesteps = 1  # Set timesteps based on the nature of your data
x_train_reshaped = np.reshape(x_train, (x_train.shape[0], timesteps, x_train.shape[1]))

input_dim = x_train_reshaped.shape[2]  # Features after reshaping

# Range of k values to test
k_values = range(2, 11)

# Empty list to store validation loss for each k
val_loss_per_k = []

# Define Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Loop over different k values
for k in k_values:
    print(f"Training with k = {k} folds...")
    
    # KFold Cross-Validation
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    
    fold_val_losses = []
    
    for train_index, val_index in kf.split(x_train_reshaped):
        # Split data into train and validation sets for the current fold
        x_train_fold, x_val_fold = x_train_reshaped[train_index], x_train_reshaped[val_index]
        
        # Define the LSTM Autoencoder model for each fold
        model = Sequential([
            LSTM(128, activation='relu', input_shape=(timesteps, input_dim), return_sequences=False),
            RepeatVector(timesteps),
            LSTM(128, activation='relu', return_sequences=True),
            TimeDistributed(Dense(input_dim))  # Output layer to match the original feature space
        ])
        
        # Compile the model model
        model.compile(optimizer='adam', loss=MeanSquaredError())
        
        # Train the model on the training fold and validate on the validation fold
        history = model.fit(x_train_fold, x_train_fold,
                                  epochs=100,
                                  batch_size=8,
                                  validation_data=(x_val_fold, x_val_fold),
                                  callbacks=[early_stopping],
                                  shuffle=True,
                                  verbose=0)
        
        # Get the validation loss for the last epoch
        val_loss = min(history.history['val_loss'])
        fold_val_losses.append(val_loss)
    
    # Average the validation loss across all folds
    avg_val_loss = np.mean(fold_val_losses)
    val_loss_per_k.append(avg_val_loss)
    print(f"Avg validation loss for k={k}: {avg_val_loss}")
    
# Find the best k-value
best_k = k_values[np.argmin(val_loss_per_k)]
print(f"Best k-value: {best_k} with lowest validation loss: {min(val_loss_per_k)}")

# Now train the final model using the best k-value
kf = KFold(n_splits=best_k, shuffle=True, random_state=42)

for train_index, val_index in kf.split(x_train_reshaped):
    x_train_fold, x_val_fold = x_train_reshaped[train_index], x_train_reshaped[val_index]
    
    # Define and train the final LSTM model
    model = Sequential([
        LSTM(128, activation='relu', input_shape=(timesteps, input_dim), return_sequences=False),
        RepeatVector(timesteps),
        LSTM(128, activation='relu', return_sequences=True),
        TimeDistributed(Dense(input_dim))
    ])
    
    model.compile(optimizer='adam', loss=MeanSquaredError())
    
    # Train on the entire training data with the best k splits
    model.fit(x_train_fold, x_train_fold,
                    epochs=100,
                    batch_size=64,
                    validation_data=(x_val_fold, x_val_fold),
                    callbacks=[early_stopping],
                    shuffle=True)

# Final evaluation on test set
x_test_reshaped = np.reshape(x_test, (x_test.shape[0], timesteps, x_test.shape[1]))
test_loss = model.evaluate(x_test_reshaped, x_test_reshaped)
print(f"Test loss after training with best k = {best_k}: {test_loss}")
```

```python
from sklearn.metrics import classification_report, accuracy_score
import numpy as np



# Calculate reconstruction errors for the training data
train_reconstructions = model.predict(x_train_reshaped)
train_mse = np.mean(np.power(x_train_reshaped - train_reconstructions, 2), axis=(1, 2))

# Set the threshold as a value (you can also tune this based on your needs)
threshold = np.percentile(train_mse, 95)  # Example: Using the 95th percentile of training errors

# Calculate reconstruction errors for the test data
test_reconstructions = model.predict(x_test_reshaped)
test_mse = np.mean(np.power(x_test_reshaped - test_reconstructions, 2), axis=(1, 2))

# Predict anomaly labels based on the threshold
y_pred = (test_mse > threshold).astype(int)  # 1 = Anomaly, 0 = Normal

# Assuming you have ground truth labels for the test set (1 for anomaly, 0 for normal)
# Replace y_test with the actual labels of your test set
 # Ground truth labels

# Calculate the accuracy and classification report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Generate classification report
report = classification_report(y_test, y_pred, target_names=["Normal", "Anomaly"])
print(report)
```
